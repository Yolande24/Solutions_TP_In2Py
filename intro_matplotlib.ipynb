# INTRODUCTION  A MATPLOTLIB

  ># Définition
>Matplotlib est une bibliothèque de traçage de graphes en python qui sert d'utilitaire de visualisation. Il a été créé par John D. Hunter en 2003. 
Matplotlib est open source et est principalement écrit en python, quelques segments sont écrits en C, Objective-C et Javascript pour la compatibilité de la plate-forme.

# Les fonctions usuelles de matplotlib

> ## .plot() 
>>-Cette fonction est utilisée pour tracer des points dans un diagramme. \
-Par défaut elle trace une ligne d'un point à un autre. \
>-Le paramètre marker de la fonction plot permet de marquer les points du diagramme,ces valeurs possibles sont : **'o' , '*' , 'D' , '+' , '1' ....**
> ## .show()
>>Elle permet d'afficher à l'écran le ou les diagramme(s) tracé(s) par la fonction plot().

> ## .savefig()
>>Elle permet de sauvegarder une copie du diagramme dans le répertoire local. 
>>>**-plt.savefig('myplot.png', dpi=300) \
>>>-plt.savefig('myplot.pdf')**

```python
import matplotlib.pyplot as plt

plt.plot([0, 6], [0, 250])
plt.savefig('myplot.jpg', dpi=300)
plt.savefig('myplot.pdf')
plt.show()
```

# Paramètres et valeurs de plot()

>Outre le paramètre marker on a aussi d'autres comme :
 >>**ms** : la taille des points; Les valeurs possibles sont les entiers \
 >**mfc** : la couleur de fond du point; \
 >**mec** : la couleur de contour du point; 
 >Les valeurs possibles sont **'g', 'b', 'r', 'c'** \
 >**linestyle** : permet de permet de préciser le style de ligne ;
ces valeurs possibles sont : **'dotted', 'dashed', 'solid', '-.' , .....**

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 1, 2, 3])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints,ypoints, marker = 'o', mec = 'b', ms = 20, mfc = 'r', linestyle = '-.') 
plt.show()
```

# Les labels

>La fonction **title()**  permet de titrer le diagramme 



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Pulsation Moyenne")
plt.ylabel("Calorie Brullées")

plt.show()
```

# Les grilles du diagramme
>Pour ajouter les lignes de grille, on utilise la fonction **grid()** \
> Le paramètre **axis** permet de spécifierla ligne de grille à afficher. Ces valeurs sont: **'x', 'y' et 'both'**. \
La valeur par défaut affiche les lignes de grille des deux axes \
> Le paramètre **color** sert à spécifier la couleur de la grillen \
> Le paramètre **linestyle** est utilser pour choisir le style des lignes de grille \
> Le paramètre **linewidth** donne la taille des lignes de grille

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis='x', color='green', linestyle = '--', linewidth = 0.5)

plt.grid(axis='y', color='red', linestyle = '--', linewidth = 1)

plt.show() 
```

# Les sous-diagrammes  -- subplot
>On peut représenter plusieurs à la fois sur un même schéma. 
>>La fonction **pyplot.subplot()** permet de définir le positionnement d'un graphe sur le rendu du schéma \
    -Elle prend trois paramètres **(row, column, number)** : \
    -Ex : **plt.subplot(1, 2, 1)** veut dire : la figure a **1 ligne, 2 colonnes et est à la position 1 à l'affichage**
    

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title("tracé 1")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)
plt.title("tracé 2")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)
plt.title("tracé 3")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.title("tracé 4")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)
plt.title("tracé 5")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)
plt.title("tracé 6")

plt.suptitle("TRACE")
plt.show()
```

# Les scatters -- Les nuages de points
 
 >On réalise les nuages de points en utilisant la fonction **pyplot.scatter()**. 
 >>Elle prend en paramètre les deux listes de nombres. elle peut aussi prendre la couleur. \
 >>Le paramètre **alpha** est utilisé pour ajuster  la transparence des points du nuage \
 >>Le paramètre **s** est utilisé pour donner la taille des points du nuage

```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = "blue",  s=100, alpha=0.5)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = "green",  s=150, alpha=0.5)

plt.show()
```

# Les Bandes
 >-Au lieu de tracer les points la fonction **bar()** va plutôt des bandes celon dont les hauteurs dépendent de la valeur des ordonnées. Elle prend en paramètre : 
 >>-(au moins deux listes) Les deux listes : **x et y** \
 >>-la couleur des bandes : **color = "r"** 
 
 >-La fonction **barh()** permet d'afficher le diagramme horizontalement

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "#4CAF50")

plt.savefig('myplot.jpg', dpi=300)
plt.savefig('myplot.pdf')
plt.show()
```

# Diagramme circulaire
>-La fonction **pyplot.pie()** permet de créer un diagramme circulaire \
>-Elle peut prendre en paramètre : 
>>-une liste d'entiers : y = [35, 25, 25, 15] \
>>-une étiquette (de type liste de str):  **labels = ["Apples","Bananas","Cherries","Dates"]** \
>>-la couleur (de type liste de str): **colors = ["black", "hotpink", "b", "#4CAF50"]**

>-En plus peut s'appliquer la légende par la fonction **pyplot.legend()** 
>>-Elle prendre en paramètre le titre de la légende : **pyplot.legend(title = "Les Fruits")**

```python

 import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

colors = ["y", "hotpink", "r", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = colors)
plt.legend(title = "Les Fruits")
plt.savefig('myplot.png')
plt.show() 
```

```python
def f(x,k):
    return 1/(1+np.exp(-k*x))

a,b = -5,5

X = np.linspace(a,b)

for k in range(1,5):
    Y = f(X,k)
    plt.plot(X,Y, label="k={}".format(k))
    
plt.legend()
plt.show()
```
